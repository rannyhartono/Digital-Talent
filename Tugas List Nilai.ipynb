```python
print("TUGAS 02: LIST NILAI")
print("Ranny Hartono - B")
print () 

import random

# GENERATE RANDOM
nilaiTugas = []
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))

print("Nilai Tugas")
print(nilaiTugas)

# ganjil
print ()
print ("Nilai Ganjil")
nilaiGanjil = []
for n in nilaiTugas:
    if n % 2 == 1: 
        nilaiGanjil.append(n)
print (nilaiGanjil)
print ()
    
# nilai > 70
nilai70 = []
for n in nilaiTugas:
    if n > 70:
        nilai70.append(n)
print("Jumlah nilai yang lebih besar dari 70 adalah sebanyak",len(nilai70))        
print ()

# standar deviasi
total = 0
for n in nilaiTugas:
    total = total + n 
average = total / len(nilaiTugas)

jumlah = 0
for n in nilaiTugas:
    a = (n - average)**2
    jumlah = jumlah + a
stdev = ( jumlah/(len(nilaiTugas)-1))**0.5

print("Standar Deviasi =", stdev)
print ()

# maksimum
maksimum = 0
for n in nilaiTugas:
    if (n>maksimum):
        maksimum = n
print("Nilai maksimum dari list nilai ini adalah",maksimum)

# minimum
minimum = 0
for n in nilaiTugas:
    if (n<minimum):
        minimum = n
print("Nilai minimum dari list nilai ini adalah",minimum)
print ()

# modus
nilaimaks=0
modus={}
for i in nilaiTugas:
    count= nilaiTugas.count(i)
    modus.update({i:count})

for key in modus:
    if modus[key]>nilaimaks:
        nilaimaks=modus[key]

for key in modus:
    if modus[key]==nilaimaks:
         print("Nilai modus dari list nilai ini adalah", key)
```
